```python
f_v=data.drop("Weight",axis=1)
c_l=data[["Weight"]]
```

```python
f_v
```

```python
c_l
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(f_v,c_l,train_size=0.8,stratify=data["Species"],random_state=100)
```

```python
species={"Perch":1,"Bream":2,"Roach":3,"Pike":4,"Smelt":5,"Parkki":6,"Whitefish":7}
```

```python
x_train["Species"]=x_train["Species"].map(species)
```

```python
x_test["Species"]=x_test["Species"].map(species)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
nrm=MinMaxScaler()
```

```python
x_train_transformed=pd.DataFrame(nrm.fit_transform(x_train),columns=x_train.columns,index=x_train.index)
```

```python
x_train_transformed
```

```python
x_test_transformed=pd.DataFrame(nrm.transform(x_test),columns=x_test.columns,index=x_test.index)
```

```python
x_test
```

```python
x_test_transformed
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train_transformed,y_train)
```

```python
predict=model.predict(x_test_transformed)
```

```python
import joblib
```

```python
joblib.dump(model,r"C:\Users\koush\OneDrive\Desktop\Moses k\Innomatics\Data\fish_model.joblib")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,predict)
```

```python
mean_squared_error(y_test,predict)
```
